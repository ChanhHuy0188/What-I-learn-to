```python
import numpy as np
import pylab as plt
import pandas as pd 
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
load_iris = datasets.load_iris()
iris_df=pd.DataFrame(load_iris.data,columns=[load_iris.feature_names])
iris_df.head()
```

**Check the Shape of Iris Dataset**

```python
load_iris.data.shape
```

**Standardization**

It is always good to standarization the data to keep all feature of data on same scale.

```python
standardizated_x=StandardScaler().fit_transform(load_iris.data)
standardizated_x[:2]
```

```python
standardizated_x.T
```

**Compute Covariance Matrix**

```python
convariance_matrix_x=np.cov(standardizated_x.T)
convariance_matrix_x
```

**Compute Eigenvalues and Eigenvectors from Covariance Matrix**

```python
eigenvalues, eigenvectors=np.linalg.eig(convariance_matrix_x)
```

```python
eigenvalues
```

```python
eigenvectors
```

**Check variance in Eigenvalues**

```python
total_of_eigenvalues = sum(eigenvalues)
variance=[(i/total_of_eigenvalues)*100 for i in sorted(eigenvalues,reverse=True)]

variance
```

**From the above result of variance**
- 1st Component=72.96%
- 2nd Component=22.85%
- 3rd Component=3.5%
- 4th Component=0.5%

So, here 3rd and 4th Components have very low variance respectively. These can be dropped. Because these components can not add any value.

**Taking 1st and 2nd Components inly and reshaping**

```python
eigenpairs=[(np.abs(eigenvalues[i]),eigenvectors[:,i]) for i in range(len(eigenvalues))]
             
#Sorting from Higher values to lower value
eigenpairs.sort(key=lambda x:x[0],reverse=True)
eigenpairs
```

```python
matrix_weighing = np.hstack((eigenpairs[0][1].reshape(4,1),eigenpairs[1][1].reshape(4,1)))
matrix_weighing
```

```python
Y= standardizated_x.dot(matrix_weighing)
Y
```

```python
plt.figure()
target_names=load_iris.target_names
y=load_iris.target
for c,i,target_name in zip('rgb',[0,1,2],target_names):
  plt.scatter(Y[y==i,0],Y[y==i,1], c=c ,label=target_name)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.title('PCA')
plt.show()
```

**Calcalating the Mean**

```python
def get_mean(value):
  total = sum(value)
  length=len(value)
  mean=total/length
  return mean
```

**Calculating the Variance**

```python
def get_variance(value):
  mean=get_mean(value)
  mean_difference_square=[pow((item-mean),2) for item in value]
  variance= sum(mean_difference_square)/float(len(value)-1)
  return variance
```

**Calcalating the Covariance**


```python
def get_covariance(value1,value2):
  value1_mean=get_mean(value1)
  value2_mean=get_mean(value2)
  values_size=len(value1)
  covariance=0.0
  for i in range(0,values_size):
    covariance+=(value1[i]-value1_mean)*(value2[i]-value2_mean)

  return covariance / float(values_size -1)
```

**Implementing a Linear Regression**

```python
def linear_regression(df):
  X=df['SqFt']
  Y=df['Price']
  m=len(X)
  
  square_feet_mean=get_mean(X)
  price_mean=get_mean(Y)

  square_feet_variance=get_variance(X)
  price_variance=get_variance(Y)
  
  covariance_of_price_and_square_feet = get_covariance(X,Y)
  w1=covariance_of_price_and_square_feet/ float(square_feet_variance)
  w0=price_mean - w1 *square_feet_mean

  #prediction --> Linear Equation
  predition = w0 + w1*X
  return predition
```

**Calling the Linear Regression Method**

```python
prediction = linear_regression(df)
```

```python
df['price (prediction)'] = prediction
df.head()
```
